**Function Python**

Una función te permite definir un bloque de código reutilizable que se puede ejecutar muchas veces dentro de tu programa

def nombre_funcion(vamos a recibir los parametros necesarios): bloque de código

```python
#funcion sinparametros o retorno de valores 
def dihola():
    print("Hola programador")
```

```python
# LLamada a la función dihola
dihola()
```

```python
# function que reciba un parametro
def holaConNombre(nombre):
    print("Hola", nombre)
```

```python
# Ingresar un nombre por teclado
nombre1 = input("Ingrese el nombre del estudiante: ")

# Llamando a la función holaConNombre
holaConNombre(nombre1)
dihola()
```

```python
# function que reciba dos parametros enteros por teclado y los multiplique
def multiplication(n1, n2):
    mul = n1 * n2
    return mul
```

```python
n1 = int(input("Ingrese el primer número: "))
n2 = int(input("Ingrese el segundo número: "))
multiplication(n1, n2)
```

```python
# Crear una función en la cual reciba tres parametros y verifique cual de ellos es el número mayor
def mayor(n1, n2, n3):
    if (n1 > n2 and n1 > n3):
        print("El número mayor es ", n1)
    else:
        if (n2 > n1 and n2 > n3):
            print("El numero mayor es ", n2)
        else:
            print("El número mayor es n3", n3)
```

```python
n1 = int(input("Ingrese el primer numero: "))
n2 = int(input("Ingrese el segundo numero: "))
n3 = int(input("Ingrese el tercer numero: "))
mayor(n1, n2, n3)
```

<h1>Que es un vector</h1>

```python
# definio el vector vacio
lista = []
tamanio = int(input("Ingese el tamaño del vector"))
for i in range(tamanio):
    n=int(input("Ingese uns valor entero"))
    lista.append(n)

# Imprima lo que tiene lista
print(lista)
```

```python
# Crear un funcion donde se ingrese o se genere un vector con número aleatoros del 0 al 10
import random as rd

def crearVector(n):
    vec = []
    for i in range(n):
        num = rd.randint(0, 10)
        vec.append(num)
    return vec

n = int(input("Ingrese el tannio del vector"))
crearVector(n)
```

```python
# CRear un functon que reciba el tamaño de un vector e ingresar el vector por teclad
def crearVec(n):
    vec = []
    for i in range(n):
        n=int(input("Ingrese los valores del vector"))
        vec.append(n)
    return vec
```

```python
n=int(input("Ingrese el tamaño del vector"))
crearVec(n)
```

```python
# Crear un funcion donde lean un vector y funcion donde encuentre el numero mayor
def leerVector(n):
    vector=[] 
    for i in range(n):
        elemento=int(input("Ingrese el valor( i + 1 ) del vector"))
        vector.append(elemento)
    return vector

def mayor(vector):
    maximo=vector[0]
    for i in vector:
        if i > maximo:
            maximo = i
    return maximo
```

```python
tamanio = int(input("Ingrese el tamanio del vector"))
vector = leerVector(tamanio)
maximo = mayor(vector)
print("Vector :", vector)
print (" El numero mayor es ", maximo)
```
