2. Dada una lista con elementos duplicados, escribir un programa que muestre una nueva lista
con el mismo contenido que la primera pero sin elementos duplicados. Para este ejercicio, no
puedes hacer uso de objetos de tipo ‘Set’.

```python
x = [2, 3, 5, 5, 9,9 , 10 ,2 ,1]
y = []

for numero in x: #bucle para recorrer la primeralista
    if numero not in y: #condicion para queen caso de que si serepite un numero no lo ponga lalista 
        y.append(numero)# con append hacemos que enla variable y se agregue la nueva lista sin duplicados

print("Original:", x)
print("sin duplicados:", y)
```
